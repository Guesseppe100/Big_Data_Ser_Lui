```python
from airflow.decorators import dag, task  # importa los decoradores para definir DAGs y tareas
from datetime import datetime  # para crear objetos datetime usados en la configuración

# Definimos el contenedor del flujo (DAG)
@dag(
    dag_id='first_dag',  # identificador único del DAG
)
def first_dag():  # función que construye las tareas y devuelve el DAG

    # Definimos la Tarea 1
    @task.python  # convierte la función siguiente en una tarea administrada por Airflow
    def first_task():
        print("This is my first task")  # acción de la tarea: imprimir saludo

    # Definimos la Tarea 2
    @task.python  # convierte la función siguiente en otra tarea de Airflow
    def second_task():
        print("This is my second task")  # acción de la tarea: imprimir despedida
    
    # Definimos la Tarea 3
    @task.python  # convierte la función siguiente en otra tarea de Airflow
    def third_task():
        print("This is my third task")  # acción de la tarea: imprimir despedida    


    # Defining task dependencies
    first = first_task()
    second = second_task()
    third = third_task()
    
    first >> second >> third

    # Establecemos el orden: primero saludar, luego despedirse
    first_task() >> second_task() >> third_task()  # operador >> define dependencia: first_task -> second_task -> third_task

# Activamos/instanciamos el DAG para que Airflow lo registre
mi_flujo = first_dag()
```
