```python
import csv
import random
from datetime import datetime, timedelta

# Output file name
output_file = "Archivo finanzas_mes_9.csv"

# Number of records
num_records = 100100

# Date range for 2024-01
start_date = datetime(2024, 1, 1)
end_date = datetime(2024, 1, 31)
date_range_days = (end_date - start_date).days

with open(output_file, mode="w", newline="", encoding="utf-8") as file:
    writer = csv.writer(file)

    # Header
    writer.writerow(["id", "salario", "gastos", "fecha", "correo"])

    for _ in range(num_records):
        # Generate values similar to your sample
        record_id = float(random.randint(1, 2000))
        salario = float(random.randint(1000, 9000))
        gastos = float(random.randint(100, 5000))

        # Random date in January 2024
        random_date = start_date + timedelta(days=random.randint(0, date_range_days))

        # Mix date formats like in your example
        if random.choice([True, False]):
            fecha = random_date.strftime("%d/%m/%Y")
        else:
            fecha = random_date.strftime("%Y-%m-%d")

        correo = f"usuario{random.randint(1000, 9999)}@universidad.edu"

        writer.writerow([record_id, salario, gastos, fecha, correo])

print(f"File '{output_file}' created successfully with {num_records} records.")
```
